```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.info()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill")
sns.catplot(data=tips, x="day", y="total_bill")
```

```python
sns.stripplot(data=tips, x="day", y="total_bill")
sns.catplot(data=tips, x="day", y="total_bill", col="time")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", col="time", kind="swarm")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", col="time", jitter=False)
```

```python
sns.catplot(data=tips, x="day", y="total_bill", col="time", kind="swarm", hue="sex")
```

```python
sns.catplot(x="day", y="total_bill", col="time", kind="swarm", hue="sex",
            data=tips.query("size != 3"))
```

```python
sns.catplot(data=tips, x="smoker", y="tip")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"])
```

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time")
```

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```
