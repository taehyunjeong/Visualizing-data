```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
df.head()
```

```python
sns.distplot(df["x"])
sns.distplot(df["y"])
```

```python
sns.jointplot(df["x"], df["y"])
```

```python
sns.jointplot(x="x", y="y", data=df)
```

```python
sns.jointplot(x="x", y="y", data=df, kind="hex")
```

```python
sns.jointplot(x="x", y="y", data=df, kind="hex", color="k")
```

```python
sns.jointplot(x="x", y="y", data=df, kind="kde")
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df["x"], df["y"])
sns.rugplot(df["x"])
sns.rugplot(df["y"], axis="y")
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df["x"], df["y"])
sns.rugplot(df["x"])
sns.rugplot(df["y"], axis="y", color="b", vertical=True)
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df["x"], df["y"])
sns.rugplot(df["x"])
sns.rugplot(df["y"], axis="y", color="b", vertical=True)
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df["x"], df["y"], cmap=cmap)
sns.rugplot(df["x"])
sns.rugplot(df["y"], axis="y", color="b", vertical=True)
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df["x"], df["y"], cmap=cmap, n_levels=60, shade=True)
sns.rugplot(df["x"])
sns.rugplot(df["y"], axis="y", color="b", vertical=True)
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df["x"], df["y"], cmap=cmap, n_levels=100, shade=True, thresh=0)
sns.rugplot(df["x"])
sns.rugplot(df["y"], axis="y", color="b")
```

```python
g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```
