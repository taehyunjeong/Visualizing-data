```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
diamonds.info()
```

```python
sns.scatterplot(data=diamonds, x="carat", y="price")
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
plt.figure(figsize=(6.5, 6.5))
sns.scatterplot(data=diamonds, x="carat", y="price")
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
plt.figure(figsize=(6.5, 6.5))
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity")
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", hue_order=clarity_ranking)
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", hue_order=clarity_ranking, size="depth")
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", hue_order=clarity_ranking, size="depth",
               palette="ch:r=-.2,d=.3_r")
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
plt.figure(figsize=(6.5, 6.5))
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", hue_order=clarity_ranking, size="depth",
               palette="ch:r=-.2,d=.3_r", sizes=(1, 8), linewidth=0)
```

```python
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```

```python
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="table",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```
