```python
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
diamonds.info()
```

```python
sns.boxenplot(data=diamonds, x="clarity", y="carat")
```

```python
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.boxenplot(data=diamonds, x="clarity", y="carat", order=clarity_ranking)
```

```python
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.boxenplot(data=diamonds, x="clarity", y="carat", order=clarity_ranking, color="b", scale="linear")
```
