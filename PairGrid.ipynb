```python
import seaborn as sns
sns.set(style="white")
```

```python
df = sns.load_dataset("iris")
df
```

```python
df.info()
```

```python
sns.PairGrid(df, diag_sharey=False)
```

```python
g = sns.PairGrid(df, diag_sharey=False)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, diag_sharey=False, hue="species")
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, diag_sharey=False, hue="species")
g.map(sns.scatterplot)
g.map_diag(sns.kdeplot) # 밀도 함수를 그려줌
```

```python
g = sns.PairGrid(df, diag_sharey=False, hue="species")
g.map_lower(sns.kdeplot) # 대각선 기준으로 아래쪽은 밀도함수
g.map_upper(sns.scatterplot) # 대각선 기준으로 위쪽은 산점도
g.map_diag(sns.kdeplot) # 밀도 함수를 그려줌
```

```python
g = sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot, lw=3)
```
