```python
import seaborn as sns
sns.set(style="white")
```

```python
df = sns.load_dataset("iris")
df
```

```python
df.info()
```

```python
sns.PairGrid(df, diag_sharey=False)
```

```python
g = sns.PairGrid(df, diag_sharey=False)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, diag_sharey=False, hue="species")
g.map(sns.scatterplot)
```
