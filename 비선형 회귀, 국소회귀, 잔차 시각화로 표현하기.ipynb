```python
import seaborn as sns
```

```python
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
 sns.lmplot(x="x", y="y", data=anscombe.query("dataset=='I'"), ci=None, scatter_kws={"s":80})
```

```python
 sns.lmplot(x="x", y="y", data=anscombe.query("dataset=='II'"), ci=None, scatter_kws={"s":80})
```

```python
 sns.lmplot(x="x", y="y", data=anscombe.query("dataset=='II'"), ci=None, scatter_kws={"s":80},
            truncate=False)
```

```python
 sns.lmplot(x="x", y="y", data=anscombe.query("dataset=='II'"), ci=None, scatter_kws={"s":80},
            truncate=False, order=2)
```

```python
 sns.lmplot(x="x", y="y", data=anscombe.query("dataset=='III'"), ci=None, scatter_kws={"s":80},
            truncate=False)
```

```python
 sns.lmplot(x="x", y="y", data=anscombe.query("dataset=='III'"), ci=None, scatter_kws={"s":80},
            truncate=False, robust=True)
```

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
tips
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips, y_jitter=.3)
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips, logistic=True, y_jitter=.03)
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips, lowess=True)
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips, lowess=False)
```
