```python
import seaborn as sns
sns.set_theme(style="ticks")
```

```python
sns.relplot(x="time", y="firing_rate", hue="coherence", col="align", data=dots, kind="line",
            size="choice", legend="full", height=5, aspect=.75, size_order=["T1", "T2"],
            facet_kws=dict(sharex=False))
```

```python
sns.relplot(x="time", y="firing_rate", hue="coherence", col="align", data=dots,
            size="choice", legend="full", height=5, aspect=.75, size_order=["T1", "T2"],
            facet_kws=dict(sharex=False))
```

```python
# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")
```

```python
# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```
