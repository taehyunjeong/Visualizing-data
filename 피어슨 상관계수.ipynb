```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
```

```python
# Generate a large random dataset
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 columns=list(ascii_letters[26:])) #import acii_letters로 col을 그려줌
d
```

```python
# Compute the correlation matrix
corr = d.corr()
corr
```

```python
# Generate a mask for the upper triangle
# 한쪽만 상관계수 나오도록 설정
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
mask
```

```python
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
cmap
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr, mask=mask)
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask)
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap="Reds")
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0)
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, vmax=.3)
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, vmax=.3, square=True, linewidths=.5)
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, vmax=.3, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, vmax=.3, square=True, linewidths=.5, 
            cbar_kws={"shrink": .5}, annot=True)
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, vmax=.3, square=True, linewidths=.5, 
            cbar_kws={"shrink": .5}, annot=True, fmt=".1f")
```
