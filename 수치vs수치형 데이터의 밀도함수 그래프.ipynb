```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
# Subset the iris dataset by species
setosa = iris.query("species == 'setosa'")
print(setosa)
virginica = iris.query("species == 'virginica'")
print(virginica)
```

```python
sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"])
# 만약 특수문자나 띄어 쓰기가 있다면 sns.kdeplot(setosa.sepal_width, setosa.sepal_length) 표현이 유리
```

```python
sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=True)
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
```

```python
sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=True)
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=True)
```

```python
sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=False)
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=False)
```

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=False)
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=False)
```

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=False)
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=False)
sns.scatterplot(data=virginica, x="sepal_width", y="sepal_length")
```

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=False, cmap="Reds")
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=False, cmap="Blues")
sns.scatterplot(data=virginica, x="sepal_width", y="sepal_length")
```

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

ax = sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=False, cmap="Reds")
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
ax = sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=False, cmap="Blues")
sns.scatterplot(data=virginica, x="sepal_width", y="sepal_length")

ax.text(2.5, 8.2, "virginica", size=16, color="blue")
ax.text(3.8, 4.5, "setosa", size=16, color="red")
```

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

ax = sns.kdeplot(setosa["sepal_width"], setosa["sepal_length"], shade=True, shade_lowest=False, cmap="Blues")
sns.scatterplot(data=setosa, x="sepal_width", y="sepal_length")
ax = sns.kdeplot(virginica["sepal_width"], virginica["sepal_length"], shade=True, shade_lowest=False, cmap="Reds")
sns.scatterplot(data=virginica, x="sepal_width", y="sepal_length")

# Add labels to the plot
ax.text(2.5, 8.2, "virginica", size=16, color="red")
ax.text(3.8, 4.5, "setosa", size=16, color="blue")
```

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))

# Draw the two density plots
ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length,
                 cmap="Reds", shade=True, shade_lowest=False)
ax = sns.kdeplot(virginica.sepal_width, virginica.sepal_length,
                 cmap="Blues", shade=True, shade_lowest=False)

# Add labels to the plot
red = sns.color_palette("Reds")[-2]
blue = sns.color_palette("Blues")[-2]
ax.text(2.5, 8.2, "virginica", size=16, color=blue)
ax.text(3.8, 4.5, "setosa", size=16, color=red)
```
