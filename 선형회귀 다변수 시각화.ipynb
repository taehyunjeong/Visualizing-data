```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
```

```python
sns.regplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", truncate=False, col="day")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", truncate=False, col="day", col_wrap=2)
```

```python
tips.corr()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", truncate=False)
```

```python
tips.groupby(["smoker"]).corr()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
```
